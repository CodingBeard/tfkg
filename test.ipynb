```python
import tensorflow as tf
import tensorflow.keras as k
print(tf.__version__)
```

```python
tf.debugging.experimental.enable_dump_debug_info(
  "debug_dump"
)

model = k.Sequential([
  k.layers.Input(name="test_input", shape=(1,)),
  k.layers.Dense(1, name="test_output")
])
model.compile()

model.save("/tfkg/base_model")
```
